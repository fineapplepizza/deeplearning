```python
#사용라이브러리
from tensorflow.keras.models import Sequential #신경망 생성
from tensorflow.keras.layers import Dense #퍼셉트론으로 구성된 신경망층 정의
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn import datasets
```

```python
#wine 데이터셋 가져오기
wine = datasets.load_wine()
```

```python
# 정답 데이터 만들기
y = wine['target']
```

```python
y=pd.get_dummies(y)
```

```python
#모든데이터가 제대로 있는지 확인하면됨
```

```python
X.shape
```

```python
# 모델 설계 
model = Sequential()
# 입력층
# input_dim = 입력 데이터의 특성 갯수
# 입력층에는 30개 뉴런을 생성할거야
#뉴런같은 경우는 모든 층마다 있음

model.add(Dense(30, input_dim = 13 , activation = 'relu'))


#은닉층
#신경망속에서 뉴런이 총 12개,#몇개 들어오는지 알 필요가 없음
model.add(Dense(12, activation='relu')) 
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='relu'))

#출력층
#내가 마지막에 출력하는 데이터가 어떤것인지 출력하기위해서 sigmoid를 씀
#모델의 구조를 만들어줌
model.add(Dense(3,activation = 'sigmoid')) 

#뉴런은 늘리면 학습하는데 오래걸리고 줄이면 학습을 제대로 못하기때문에
#적당하게 쌓아야한다.
#뉴런의 갯수를 조금 줄여가는게 좋다 
#분류데이터이기 때문에 답이 마지막에 1개로 주어짐 1개를 출력하기 위해
#출력층에서는 한 개의 결과값을 출력해야함
#처음에는 다양한 방법으로 분석해보고 점점 줄여나가면서 한 개의 결과 값을 출력
# 중간 중간에 키웠다가 줄였다가 해도 상관 없지만 은닉층 마지막부분에서는 뉴런의
#수를 줄여나가는 게 좋습니다.
# 결과값과 출력층은 동일하게 하는게 좋다.
#와인데이터의 정답이 총 3개 이기때문에 3개의 출력갯수로 하면 1번와인이 정답확률,2번와인 정답확률,3번와인 정답확률로 출력을 함
#모델이 출력하는 갯수와 다르게 나타나면 오류가 뜸
```

```python
# 모델 컴파일
# 설계한 모델을 실행할수 있도록 다양한 정보 입력
# loss: 손실함수(손실함수가 작을 수록 잘 설계된 모델로봄),오차
# optimizer : 경사하강법 종류(어떤 방법을 통해서 분석할 것이냐)
# metrics : 평가방법
#binary_crossentropy:카테고리가 2개인 경우
#categorical_crossentropy: 3개 이상의 경우
model.compile(loss = 'categorical_crossentropy',
             optimizer = 'adam',
             metrics = ['accuracy']) 
```

```python
# 위에서 뉴런을 쌓은 것은 생각하는 방법이라고 할 수 있다.

# 모델 학습
# 문제,정답, 반복할 횟수,한번에 확인할 데이터 수
# batch_size 데이터를 분석 할 때 한 번에 10개씩 묶어서 분석
# batch_size 크게 잡으면 분석하는 속도가 빨라지고 작게 잡으면 속도가 오래 걸린다.
model.fit(X,y,epochs = 500, batch_size = 10)
```
