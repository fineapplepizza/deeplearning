- 470개 데이터로 구성
- 17개의 특성데이터와 1개의 라벨 데이터
- 라벨데이터 0은 사망 1은 생존

```python
import pandas as pd 
data = pd.read_csv('ThoraricSurgery.csv',header=None) #header=None 첫번째 데이터가 특성이 되는 것이 사라짐

data.head()
```

```python
# 특성 데이터와 라벨 데이터 분리
X= data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
X.shape,y.shape
```

```python
# 파라미터 초기화를 위한 랜덤 seed 설정(동일한 값으로 파라미터값이 설정)
import numpy as np
import tensorflow as tf
seed = 100
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 훈련데이터와 테스트 데이터로 분리
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=3)
```

```python
# 신경망 설계 
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Activation

model = Sequential()
#신경망의 층을 추가
model.add(Dense(units = 10, input_dim = 16, Activation='sigmoid'))

model.add(Dense(units=6, Activation='sigmoid'))
model.add(Dense(units= 3 , Activation='sigmoid'))

#출력층
model.add(Dense(units=1, Activation='softmax'))
```
