```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
X = iris['data']
y = iris['target']
```

```python
X # np.array형식으로 나와도 딥러닝에서는 돌리는데 상관이없다.
```

```python
# train과 test로 분리
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=10)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# 데이터 정제 완료
# 데이터 입출력 확인
# 입력 데이터 수 알아보기
# input_dim = 4(입력해야 될 데이터 수)
X_train
```

```python
# 마지막 층에서 출력할 데이터 수 알아보기
# units = 3
y_train
```

### 딥러닝 모델 구성하기


```python
from tensorflow.keras import Sequential # 딥러닝 모델 생성 ,딥러닝 모델의 뼈대
from tensorflow.keras.layers import Dense# 층을 쌓는다.
```

```python
# 1. 딥러닝 모델 생성
model = Sequential() #생성한 딥러닝 모델은 뼈대가 된다.

# 입력층(출력층의 개수(units),입력층의 개수(input_dim),활성화함수(activation))
model.add(Dense(8, input_dim=4, activation='sigmoid'))

# 중간층(은닉층) - 마음대로 늘려주고 줄여줘도 상관없다.
#
model.add(Dense(4, activation = 'sigmoid'))
model.add(Dense(3, activation = 'sigmoid'))
#층이 깊지않아서 sigmoid해도 크게 상관없지만 
#마지막 층은 반드시 softmax로 바꿔줘야함
#마지막층을 sigmoid로 쓰면 전체확률이 100퍼센트가 넘어가서 판단이 불가

# 출력층
# 출력층의 개수(units),활성화함수(activation)
# 마지막층은 출력층의 개수가정해져 있음? 정답을 맞추려고 하는 타겟의 수 >3
# 추천하는 활성화함수도 존재
model.add(Dense(3,activation='softmax'))
```

```python
model.summary()
```

```python
# 2. 모델에 학습 방법 설정
# loss: 손실함수
# metrics : 평가방법
# optimizer : 경사하강법의 종류
model.compile(loss='categorical_crossentropy', #다중분류일때 사용
              optimizer = 'adam',
              metrics = ['accuracy']
             )
```

```python
# 3. 모델 학습
# 정답 데이터를 변형 
y_train_one_hot = pd.get_dummies(y_train)
y_test_one_hot = pd.get_dummies(y_test)
```

```python
# 특성데이터(X) , 라벨데이터(y), 반복할 횟수(epochs), 한번에 학습할 데이터 수(batch_size)
# 필수로 적어야 하는 것 : X,y,epochs
#batch_size는 안적어도 알아서 적당히 선택해서 돌려짐
#loss: 1.1068 - accuracy: 0.3125 # 맨 처음 학습
#oss: 0.4166 - accuracy: 0.9643 # 모델 초기화를 안 한후 학습을 시켜 전 epoch를 기억
#loss: 1.1751 - accuracy: 0.3125 # 재 학습을 시켰는데 모델 초기화 한 후
model.fit(X_train, y_train_one_hot,epochs=500)
```

```python
# 4. 모델 평가하기
model.evaluate(X_test,y_test_one_hot)
```
