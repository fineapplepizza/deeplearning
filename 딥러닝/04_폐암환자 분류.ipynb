폐암 환자 데이터셋
- 이진 분류
- 470개 데이터로 구성
- 17개의 특성데이터와 1개의 라벨 데이터
- 라벨데이터 0은 사망 1은 생존

```python
import pandas as pd 
data = pd.read_csv('ThoraricSurgery.csv',header=None) #header=None 첫번째 데이터가 특성이 되는 것이 사라짐

data.head()
```

```python
# 특성 데이터와 라벨 데이터 분리
X= data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
X.shape,y.shape
```

```python
# 파라미터 초기화를 위한 랜덤 seed 설정(동일한 값으로 파라미터값이 설정)
import numpy as np
import tensorflow as tf
seed = 100
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 훈련데이터와 테스트 데이터로 분리
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=10)
```

```python
# 신경망 설계 
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
#신경망의 층을 추가
model.add(Dense(units = 20, input_dim = 17, activation='sigmoid'))

model.add(Dense(units=15, activation='sigmoid'))
model.add(Dense(units= 10 , activation='sigmoid'))
model.add(Dense(units= 5 , activation='sigmoid'))

#출력층
model.add(Dense(units=2, activation='softmax'))
model.summary()
```

```python
# 모델 학습 방법 설정
model.compile(loss='categorical_crossentropy',
             optimizer = 'adam',
             metrics=['accuracy'])
```

```python
import pandas as pd
y_train_one_hot = pd.get_dummies(y_train)
y_test_one_hot = pd.get_dummies(y_test)
```

```python
# 모델 학습
# 머신러닝에서는 학습과 검증이 따로
# 딥러닝에서는 한번에 가능
# fit 함수 속성으로 validation_data - 검증
#기존에 나온 정확도와 validation_data에서 나온 정확도가 차이가 나면 과대적합 가능성이 있다.
model.fit(X_train,y_train_one_hot,epochs= 2000,validation_data=[X_test,y_test_one_hot])
```

```python
#train보다 test가 더 높으면 일반화가 잘됨
```

```python
model.evaluate(X_train,y_train)
```

```python
#모델 평가
#오차[0]와 정확도[1]
model.evaluate(X_train,y_train_one_hot)
```
